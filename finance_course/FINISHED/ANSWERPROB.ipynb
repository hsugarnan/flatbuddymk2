```python
import pandas as pd

# Create a DataFrame to represent the possible sums of rolling three dice (3 to 18).
sums = pd.DataFrame(index=range(3, 19))

# Assign probabilities to each possible sum based on the number of combinations that result in that sum.
# Note: These values are pre-calculated based on combinations of three dice.
sums['Prob'] = [1, 3, 6, 10, 15, 21, 25, 27, 27, 25, 21, 15, 10, 6, 3, 1]

# Convert these counts to probabilities by dividing each by 216 (since 6 * 6 * 6 = 216 possible outcomes).
sums['Prob'] = sums['Prob'] / 216

# Calculate the mean and variance for the distribution
mean = pd.Series(sums.index * sums['Prob']).sum()
var = pd.Series(((sums.index - mean)**2) * sums['Prob']).sum()

# Print the results
print("Mean of the distribution:", mean)
print("Variance of the distribution:", var)
```
