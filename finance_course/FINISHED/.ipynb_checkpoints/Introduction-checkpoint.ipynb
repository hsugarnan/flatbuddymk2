```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats



# Test pandas
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
print("\nPandas DataFrame:")
print(df)

# Test numpy
array = np.array([1, 2, 3])
print("\nNumPy array:")
print(array)

# Test matplotlib
plt.plot(array, label="Line plot")
plt.title("Matplotlib Test Plot")
plt.legend()
plt.show()

# Test seaborn
sns.set(style="whitegrid")
sns.barplot(x='A', y='B', data=df)
plt.title("Seaborn Test Plot")
plt.show()

# Test scipy
sample_data = np.random.normal(size=100)
statistic, p_value = stats.shapiro(sample_data)
print("\nSciPy Shapiro-Wilk Test:")
print(f"Statistic: {statistic}, p-value: {p_value}")
```
