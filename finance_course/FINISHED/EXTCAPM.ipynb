```python
import pandas as pd
import matplotlib.pyplot as plt

# Assume these values have been calculated
risk_free_rate = 0.015  # 1.5% risk-free rate
market_return = 0.087  # Example market return (8.7%)
beta = 2.0  # Example beta for Tesla (replace with actual value)
expected_return_tesla = risk_free_rate + beta * (market_return - risk_free_rate)

# Creating a DataFrame for visualization
data = {
    'Rate Type': ['Risk-Free Rate', 'Market Return', 'Tesla Expected Return'],
    'Value': [risk_free_rate, market_return, expected_return_tesla]
}
returns_df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(returns_df['Rate Type'], returns_df['Value'], color=['blue', 'orange', 'green'])
plt.ylabel('Return')
plt.title('Comparison of Returns')
plt.ylim(0, 0.2)  # Setting the y-axis limit for better visibility
plt.axhline(y=risk_free_rate, color='blue', linestyle='--', label='Risk-Free Rate (1.5%)')
plt.axhline(y=market_return, color='orange', linestyle='--', label='Market Return (8.7%)')
plt.axhline(y=expected_return_tesla, color='green', linestyle='--', label='Tesla Expected Return')

# Adding a legend
plt.legend()
plt.grid(axis='y')
plt.show()

```
